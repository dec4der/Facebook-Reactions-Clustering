```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\abhig\Desktop\ML_Codebasics\my datasets\facebook.csv")
df
```

```python
df.status_type.unique()
```

```python
st = pd.get_dummies(df.status_type, drop_first=True)
df = pd.concat([df,st], axis = 'columns')
df = df.drop("status_published", axis = "columns")
df
```

```python
df = df.drop("status_id", axis = "columns")
df
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
km = KMeans(n_clusters = 3)
km
```

```python
cmt_predicted = km.fit_predict(df[['num_reactions','num_comments']])
cmt_predicted
```

```python
df['cluster'] = cmt_predicted
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.num_reactions, df1['num_comments'], color = 'green')
plt.scatter(df2.num_reactions, df2['num_comments'], color = 'blue')
plt.scatter(df3.num_reactions, df3['num_comments'], color = 'black')

plt.xlabel('Reactions')
plt.ylabel('Comments')
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['num_comments']])
df[['num_comments']] = scaler.transform(df[['num_comments']])

scaler.fit(df[['num_reactions']])
df[['num_reactions']] = scaler.transform(df[['num_reactions']])

df
```

```python
km = KMeans(n_clusters = 3)
cmt_predicted = km.fit_predict(df[['num_reactions','num_comments']])
cmt_predicted
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.num_reactions, df1['num_comments'], color = 'green')
plt.scatter(df2.num_reactions, df2['num_comments'], color = 'blue')
plt.scatter(df3.num_reactions, df3['num_comments'], color = 'black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='purple', marker = '*', label = 'centroid')

plt.xlabel('Reactions')
plt.ylabel('Comments')
plt.legend()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
  km = KMeans(n_clusters = k)
  km.fit(df[['num_reactions','num_comments']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_rng,sse)
```
